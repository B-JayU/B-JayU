## Top10 상권 알아보기

```python
data = data[['상권_코드_명','남성_매출_건수','여성_매출_건수']]
data['총 매출 건수'] = data['남성_매출_건수'] + data['여성_매출_건수']
data = data[['상권_코드_명', '총 매출 건수']]
data
```

```python
data1 = data.groupby(['상권_코드_명']).sum()
data1
```

```python
data_hap = data1.sort_values(['총 매출 건수'], ascending = False)
data_hap
```

```python
data_top = data_hap[['총 매출 건수']][:10]
data_top
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import seaborn as sns
colors = sns.color_palette('hls',len(data_top.index))

plt.figure(figsize=(10,8))
xtick_label_position = list(range(len(data_top.index)))
plt.xticks(xtick_label_position, data_top.index, rotation=45)

plt.bar(xtick_label_position, data_top['총 매출 건수'], color=colors) ## 바 차트 출력
 
plt.title('2021년 서울시 TOP10 매출건수 상권',fontsize=20) ## 타이틀 출력
plt.xlabel('상권') ## x축 라벨 출력
plt.ylabel('매출건수') ## y축 라벨 출력
plt.show()

plt.savefig("2021년 서울시 TOP10 매출건수 상권.png")
```

## 업종별 인기 성별 및 연령대 알아보기

```python
data = pd.read_csv('/content/drive/MyDrive/CJ네트웍스 Remote Internship/매출정보_상권분류코드_위치정보(위도경도)합치기_1109.csv')
```

```python
data.columns
```

```python
six_service = ['커피-음료', '편의점', '화장품', '치킨전문점', '호프-간이주점', '운동/경기용품']
```

```python
data_pop_MF = data[['서비스_업종_코드_명','남성_매출_비율', '여성_매출_비율']]
data_pop_age = data[['서비스_업종_코드_명', '연령대_10_매출_비율', '연령대_20_매출_비율', '연령대_30_매출_비율', '연령대_40_매출_비율',
       '연령대_50_매출_비율', '연령대_60_이상_매출_비율']]
```

```python
data_pop_MF = data_pop_MF[(data_pop_MF['서비스_업종_코드_명'] == six_service[0]) | (data_pop_MF['서비스_업종_코드_명'] == six_service[1]) | (data_pop_MF['서비스_업종_코드_명'] == six_service[2]) |
       (data_pop_MF['서비스_업종_코드_명'] == six_service[3]) | (data_pop_MF['서비스_업종_코드_명'] == six_service[4]) | (data_pop_MF['서비스_업종_코드_명'] == six_service[5])]
```

```python
data_pop_MF = data_pop_MF.groupby(['서비스_업종_코드_명']).mean()
```

```python
data_pop_M = data_pop_MF
data_pop_M['성별'] = '남성'
data_pop_M['성별에 따른 매출비율'] = data_pop_M['남성_매출_비율']
data_pop_M = data_pop_M[['성별', '성별에 따른 매출비율']]
data_pop_M
```

```python
data_pop_F = data_pop_MF
data_pop_F['성별'] = '여성'
data_pop_F['성별에 따른 매출비율'] = data_pop_F['여성_매출_비율']
data_pop_F = data_pop_F[['성별', '성별에 따른 매출비율']]
data_pop_F
```

```python
data_pop_MF2 = pd.concat([data_pop_M, data_pop_F])
data_pop_MF2.index
```

```python
import seaborn as sns
colors = sns.color_palette('hls',2)

plt.figure(figsize=(15,8))
sns.barplot(x=data_pop_MF2.index, y='성별에 따른 매출비율', hue='성별', data=data_pop_MF2)
plt.show()
plt.savefig("업종-성별 인기 선호도(매출비율기준).png",dpi=300)
```

### 업종-연령별 인기 선호도

```python
data_pop_age = data[['서비스_업종_코드_명', '연령대_10_매출_비율', '연령대_20_매출_비율', '연령대_30_매출_비율', '연령대_40_매출_비율',
       '연령대_50_매출_비율', '연령대_60_이상_매출_비율']]
```

```python
data_pop_age = data_pop_age[(data_pop_age['서비스_업종_코드_명'] == six_service[0]) | (data_pop_age['서비스_업종_코드_명'] == six_service[1]) | (data_pop_age['서비스_업종_코드_명'] == six_service[2]) |
       (data_pop_age['서비스_업종_코드_명'] == six_service[3]) | (data_pop_age['서비스_업종_코드_명'] == six_service[4]) | (data_pop_age['서비스_업종_코드_명'] == six_service[5])]
```

```python
data_pop_age = data_pop_age.groupby(['서비스_업종_코드_명']).mean()
data_pop_age
```

```python
data_pop_10 = data_pop_age
data_pop_10['연령대'] = '10대'
data_pop_10['연령대에 따른 매출비율'] = data_pop_10['연령대_10_매출_비율']
data_pop_10 = data_pop_10[['연령대', '연령대에 따른 매출비율']]
data_pop_10 = data_pop_10.dropna()
data_pop_10
```

```python
data_pop_10 = data_pop_age
data_pop_10['연령대'] = '10대'
data_pop_10['연령대에 따른 매출비율'] = data_pop_10['연령대_10_매출_비율']
data_pop_10 = data_pop_10[['연령대', '연령대에 따른 매출비율']]
data_pop_10 = data_pop_10.dropna()


data_pop_20 = data_pop_age
data_pop_20['연령대'] = '20대'
data_pop_20['연령대에 따른 매출비율'] = data_pop_20['연령대_20_매출_비율']
data_pop_20 = data_pop_20[['연령대', '연령대에 따른 매출비율']]
data_pop_20 = data_pop_20.dropna()

data_pop_30 = data_pop_age
data_pop_30['연령대'] = '30대'
data_pop_30['연령대에 따른 매출비율'] = data_pop_30['연령대_30_매출_비율']
data_pop_30 = data_pop_30[['연령대', '연령대에 따른 매출비율']]
data_pop_30 = data_pop_30.dropna()

data_pop_40 = data_pop_age
data_pop_40['연령대'] = '40대'
data_pop_40['연령대에 따른 매출비율'] = data_pop_40['연령대_40_매출_비율']
data_pop_40 = data_pop_40[['연령대', '연령대에 따른 매출비율']]
data_pop_40 = data_pop_40.dropna()

data_pop_50 = data_pop_age
data_pop_50['연령대'] = '50대'
data_pop_50['연령대에 따른 매출비율'] = data_pop_50['연령대_50_매출_비율']
data_pop_50 = data_pop_50[['연령대', '연령대에 따른 매출비율']]
data_pop_50 = data_pop_50.dropna()

data_pop_60 = data_pop_age
data_pop_60['연령대'] = '60대이상'
data_pop_60['연령대에 따른 매출비율'] = data_pop_60['연령대_60_이상_매출_비율']
data_pop_60 = data_pop_60[['연령대', '연령대에 따른 매출비율']]
data_pop_60 = data_pop_60.dropna()
```

```python
data_pop_age = pd.concat([data_pop_10,data_pop_20, data_pop_30, data_pop_40, data_pop_50,data_pop_60])
data_pop_age
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=data_pop_age.index, y='연령대에 따른 매출비율', hue='연령대', data=data_pop_age)
plt.show()
plt.savefig("업종-연령대별 인기 선호도(매출비율기준).png",dpi=300)
```
